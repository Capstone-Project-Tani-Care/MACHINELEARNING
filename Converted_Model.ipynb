```python
from google.colab import drive
import os
import tensorflow as tf
```

```python
# Mount Google Drive
drive.mount('/content/drive')
```

```python
# file sumber dan folder tujuan
source_files = [
    '/content/drive/My Drive/Machine Learning/CORN/DenseNet121_custom.keras',
    '/content/drive/My Drive/Machine Learning/Potato/ResNet152V2_custom.keras',
    '/content/drive/My Drive/Machine Learning/Rice/DenseNet121_custom.keras',
    '/content/drive/My Drive/Machine Learning/Soybean/DenseNet121_custom.keras',
    '/content/drive/My Drive/Machine Learning/Tomato/ResNet152V2_custom.keras'
]
destination_folder = '/content/drive/My Drive/Machine Learning/Converted_Model'

os.makedirs(destination_folder, exist_ok=True)
```

```python
def get_folder_name(file_path):
    return os.path.basename(os.path.dirname(file_path))
```

```python
def convert_keras_to_tflite(source_files, dest_folder):
    for keras_file in source_files:
        if keras_file.endswith('.keras') or keras_file.endswith('.h5'):
            try:
                # Ambil nama folder untuk nama file tujuan
                folder_name = get_folder_name(keras_file)
                tflite_file_path = os.path.join(dest_folder, f"{folder_name}.tflite")

                # Load model Keras
                print(f"Mengonversi: {keras_file} -> {tflite_file_path}")
                model = tf.keras.models.load_model(keras_file)

                # Konversi ke TFLite
                converter = tf.lite.TFLiteConverter.from_keras_model(model)
                tflite_model = converter.convert()

                # Simpan file TFLite
                with open(tflite_file_path, 'wb') as f:
                    f.write(tflite_model)

                print(f"  Berhasil: {keras_file} -> {tflite_file_path}")
            except Exception as e:
                print(f"  Gagal mengonversi {keras_file}: {e}")
```

```python
# Jalankan proses konversi
convert_keras_to_tflite(source_files, destination_folder)
print(f"Semua file telah berhasil dikonversi dan disimpan di {destination_folder}.")
```
